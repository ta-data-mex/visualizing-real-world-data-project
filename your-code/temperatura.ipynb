```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('temperatura_promedio_1985-2018.csv', encoding = "ISO-8859-1", low_memory=False)
data1
```

```python
data2 = data1.rename(columns={'Año':'anio',
                             'Clave Entidad':'clave_entidad',
                             'Entidad':'entidad',
                             'Anual':'temperatura_promedio_anual'})
```

```python
data3 = data2[['anio', 'clave_entidad', 'entidad', 'temperatura_promedio_anual']]
```

```python
dataset1 = pd.DataFrame(data3)
dataset1
```

```python
dataset1.dtypes
```

```python
dataset1['anio'] = dataset1['anio'].astype(str)
dataset1['clave_entidad'] = dataset1['clave_entidad'].fillna(0)
dataset1['clave_entidad'] = dataset1['clave_entidad'].astype(int)
dataset1['clave_entidad'] = dataset1['clave_entidad'].astype(str).str.zfill(2)
dataset1['temperatura_promedio_anual'] = dataset1['temperatura_promedio_anual'].fillna(0)
dataset1
```

```python
dataset1.dtypes
```

```python
dataset = dataset1.loc[(dataset1.anio == '2015') | (dataset1.anio == '2016') | (dataset1.anio == '2017') | (dataset1.anio == '2018')]
dataset
```
