```python
api_buy_ve = 'buy-bitcoins-online/ve/venezuela/.json'
api_sell_ve = 'sell-bitcoins-online/ve/venezuela/.json'
```

```python
response_buy_ve = requests.get(url+api_buy_ve ) #Extraigo ofertas de compra de Localbitcoins.com
buy_ve = response_buy_ve.json()
buy_ve
```

```python
response_sell_ve = requests.get(url+api_sell_ve )  #Extraigo ofertas de venta de Localbitcoins.com
sell_ve = response_sell_ve.json()
sell_ve
```

```python
list_buy_trades = buy_ve['data']['ad_list']
list_sell_trades = sell_ve['data']['ad_list']
list_sell_trades
```

```python
list_trades = list_sell_trades + list_buy_trades
```

```python
online_provider = []
currency = []
bank_name = []
max_amount_available = []
for trade in list_trades:
    online_provider.append(trade['data']['online_provider'])
    currency.append(trade['data']['currency'])
    bank_name.append(trade['data']['bank_name'])
    max_amount_available.append(float(trade['data']['max_amount_available']))
#print(online_provider)
#print(currency)
#print(bank_name)
print(max_amount_available)
```

```python
data = pd.DataFrame() #Creo dataframe
```

```python
data['online_provider'] = online_provider 
data['bank_name'] = bank_name
data['currency'] = currency
data['max_amount_available'] = max_amount_available
```
