We will explore the information regarding the Reporting Crime in Boston between Jun 2015 until Sept 2018

```python
#Fist we will import the data and explore its type of values
df0 = pd.read_csv('crime.csv', encoding='latin-1')
df0.head()
df0.info()
```

```python
#The types of crimes are classified in groups:
df0.OFFENSE_CODE_GROUP.unique()
```

```python
#There is another csv file to check the code of crime and it's description
df1 = pd.read_csv('offense_codes.csv', encoding='latin-1')
df1.head()
df1.info()
```

```python
#I wanted to change the columns name of code in order to merge it with the second csv file (Offense_Codes)
df0.rename(columns={'OFFENSE_CODE':'CODE'}, inplace=True)
df0.info()
```

```python
#First I started to suggest order the data crime by code in order to start counting it.
df2 = df0.groupby('CODE').count().sort_values('INCIDENT_NUMBER', ascending = False)
#df2 = df2.drop(columns = ['YEAR', 'MONTH', 'HOUR'])
df2 = df2.reset_index()
#df2.info()
df2.head()
```

```python
#What are the top 10 crimes groups during all the period evaluated?
df4 = df0.groupby('OFFENSE_CODE_GROUP').count().sort_values('INCIDENT_NUMBER', ascending=False).head(10)
plt.title("Top 10 Crime Types between Jun 15 - Sep 18")
plt.ylabel("# Incidents")
df4['INCIDENT_NUMBER'].plot.bar()
df4
#Top Crimes during all the period of time
```

```python
#But not all the crimes involved shootings, they were a few. Let's check when was the period with major and minor crimes involved with shootings
df5 = pd.pivot_table(df0, index= ['YEAR', 'MONTH'], columns = ['SHOOTING'], values = ['INCIDENT_NUMBER'], aggfunc={'INCIDENT_NUMBER': 'count'},  fill_value = 0)
df5

df5['INCIDENT_NUMBER'].plot(legend=False, color = "r", figsize = (10,12))
plt.title("# INCIDENTS reported involving shootings between Jun 15 - Sep 18")
plt.ylabel("# Incidents")
plt.show()

#plt.xy(figsize=(20,21))
```

```python
#Where were the districts with major number of incidents?
df6 = pd.pivot_table(df0, index= ['DISTRICT'], columns = ['YEAR'], values = ['INCIDENT_NUMBER'], aggfunc={'INCIDENT_NUMBER': 'count'},  fill_value = 0)
df6

XX = df6['INCIDENT_NUMBER'].plot()
#XX.set_xticklabels(XX.get_xticklabels(), rotation = 90)
plt.title("Districts per number of incidents")
plt.ylabel("# Incidents")
plt.show()
```

```python
#After evaluating all the districts - areas, I will explore the details of the top 3 criminilistc districts:
df6 = pd.pivot_table(df0, index= ['DISTRICT', 'REPORTING_AREA'], columns = ['YEAR'], values = ['INCIDENT_NUMBER'], aggfunc={'INCIDENT_NUMBER': 'count'},  fill_value = 0)
df6.loc[['A1','A15', 'B2']].head()
#df6
```

```python
#On the other side, I will explore the offense groups by each year
df7 = pd.pivot_table(df0, index= ['OFFENSE_CODE_GROUP'], columns= ['YEAR'], values = ['INCIDENT_NUMBER'], aggfunc={'INCIDENT_NUMBER': 'count'},  fill_value = 0)
df7.sort_values('OFFENSE_CODE_GROUP', ascending = True).head()
xy = df7['INCIDENT_NUMBER'].plot()
plt.title("Offense Group per year")
plt.ylabel("# Incidents")
xy.set_xticklabels(xy.get_xticklabels(), rotation = 90)
#plt.xy(figsize=(20,21))
plt.show()
#se puede ampliar el trabajo hacia mayor n√∫mero de incidencias en algunas calles especificas e integrarlas con los APIs de GOOGLE MAPS
```
