```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt 
import seaborn as sns

pd.set_option('display.max_colwidth', 100)
```

```python
df = pd.read_csv('Accidentes de Tránsito Terrestre en Zonas Urbanas y Suburbanas.csv')
```

```python
df.info()
```

```python
df['ALIENTO'].value_counts()
```

```python
df['CINTURON'].value_counts()
```

```python
muertes = pd.pivot_table(df,
                        values=['CONDMUERTO','PASAMUERTO','PEATMUERTO','CICLMUERTO','OTROMUERTO','NEMUERTO'], 
                        #columns=['GENERO'],
                        index=['MES'],
                        aggfunc={'CONDMUERTO': 'sum',
                                 'PASAMUERTO': 'sum',
                                 'PEATMUERTO': 'sum',
                                 'CICLMUERTO': 'sum',
                                 'OTROMUERTO': 'sum',
                                 'NEMUERTO': 'sum',
                                },
                        margins=True,
                        margins_name='Total'
                        )
muertes
```

```python
heridos = pd.pivot_table(df,
                        values=['CONDHERIDO','PASAHERIDO','PEATHERIDO','CICLHERIDO','OTROHERIDO','NEHERIDO'], 
                        #columns=['GENERO'],
                        index=['MES'],
                        aggfunc={'CONDHERIDO': 'sum',
                                 'PASAHERIDO': 'sum',
                                 'PEATHERIDO': 'sum',
                                 'CICLHERIDO': 'sum',
                                 'OTROHERIDO': 'sum',
                                 'NEHERIDO': 'sum',
                                },
                        margins=True,
                        margins_name='Total'
                        )
heridos
```

```python
serie = df.GENERO.value_counts()
serie.plot.pie(autopct='%1.1f%%');
```
