```python
try_table = arranged_table.head(34)
```

```python
l = [str(i) for i in try_table['value']['CO']]
```

```python
col1 = list(try_table.columns.levels[0])
col2 = list(try_table.columns.levels[1])
```

```python
for col in col1:
    for val in col2:
        l = [str(i) for i in try_table[col][val]]
        
        for x in range(len(l)):
                if l[x] == 'nan':
                    y = x-1
                    z = x+1
                    if l[x+1]!='nan':
                        l[x] = str((float(l[y])+float(l[z]))/2)
                    else:
                        while l[z]=='nan':
                            z +=1
                        l[x] = str((float(l[y])+float(l[z]))/2)
                else:
                    pass
        try_table[col][val] = [float(i) for i in l]
```
