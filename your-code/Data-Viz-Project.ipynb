```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ingresos_del_sistema_de_transporte_colectivo_metro.csv',parse_dates=['FECHA'], dayfirst=True)
```

```python
df.head(50)
```

```python
df.info()
```

```python
df['LINEA_12']=df['LINEA_12'].fillna(0)
```

```python
df.head()
```

```python
df['LINEA_1']=df['LINEA_1'].astype('int')
```

```python
df[['LINEA_2','LINEA_3']]=df[['LINEA_2','LINEA_3']].astype('int')
```

```python
df[['LINEA_4','LINEA_5', 'LINEA_6', 'LINEA_7', 'LINEA_8', 'LINEA_9', 'LINEA_A', 'LINEA_B', 'LINEA_12']]=df[['LINEA_4','LINEA_5', 'LINEA_6', 'LINEA_7', 'LINEA_8', 'LINEA_9', 'LINEA_A', 'LINEA_B', 'LINEA_12']].astype('int')
```

```python
df['FECHA'] = pd.to_datetime(df['FECHA'], infer_datetime_format=True)
                             #, dayfirst=True)
```

```python
df['AÑO'] = df['FECHA'].dt.year
```

```python
df['MES'] = df['FECHA'].dt.month
```
