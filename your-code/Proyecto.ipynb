# DECILES

```python
#Cargar la base
import pandas as pd
import numpy as np
```

```python
#Cargar la base
deciles = pd.read_csv("decil.csv", encoding='latin-1')  #Le agregas encoding porque está en español
```

```python
deciles
```

```python
#Transponemos por decil 
decil_limp = deciles.pivot(index="Año", columns="Decil", values = "Ingreso corriente total per cápita").reset_index(drop=True)
```

```python
decil_limp
```

```python
decil_limp.columns
```

```python
#Botamos la row de NaN
decil_limp = decil_limp.drop([0], axis=0)
decil_limp
```

```python
#Incluimos columna de años que se perdió en el Pivot Table. #PREGUNTAAAARRR A GUS
#decil_limp["Año"] = ["2008", "2010", "2012", "2014", "2016", "2018"]
decil_limp.insert(1, "Año", [2008, 2010, 2012, 2014, 2016, 2018], True)
decil_limp
#año = [2008, 2010, 2012, 2014, 2016, 2018]
#decil_limp.size
```

```python
#Botamos la columna dos
decil_limp = decil_limp.drop([decil_limp.columns[0]], axis="columns")
```

```python
#Cambiamos ordenes de columnas
decil_limp = decil_limp[["Año", 'I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX', 'X']]
decil_limp
```

# Hacemos la gráfica de deciles

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#Ajustamos el paquete
sns.set(rc={'figure.figsize':(15,8)}, style ="darkgrid") # style = "whitegrind"
```

```python
#GRAFICA INTERMEDIA 1
sns.lineplot(decil_limp.Año, decil_limp.IV)
```

```python
#GRAFICA INTERMEDIA 2
decil_limp.plot(x="Año", y=['I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX', 'X'])
```

```python
#GRAFICA INTERMEDIA 3
decil_limp_mean = decil_limp[['I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX', 'X']].agg("mean")
decil_limp_mean.plot.bar()
```

```python
#GRAFICA INTERMEDIA 4 
decil_limp_mean.plot.box()
```

```python
#GRAFICA INTERMEDIA 5
decil_limp.boxplot(column=['I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX', 'X'])
plt.xlabel("Decil")
plt.ylabel("Ingreso promedio")
plt.title("Ingreso promedio por decil") #loc="left")
plt.savefig('Boxplot Ingreso promedio por decil.png', dpi=96)
```

```python
#Intentemos que salga la gráfica de burbuja
decil_media_size = decil_limp_mean
decil_media_size = pd.DataFrame(decil_media_size).reset_index()
decil_media_size = decil_media_size.rename(columns={0: "Media_ingreso"})
decil_media_size.columns
```

```python
#Agregamos tamaño de la ENIGH
tamaño = [1.9,3.2,4.2,5.2,6.3,7.5,9.1,11.4,15.7,35.4]
decil_media_size["Tamaño"] = tamaño
```

```python
decil_media_size
```

```python
#GRAFICA FINAL 1

plt.scatter(x=decil_media_size.Decil, y=decil_media_size.Media_ingreso, c="green", s=decil_media_size.Tamaño*400, alpha=0.4, edgecolors="grey", linewidth=2)
#decil_limp.boxplot(column=['I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX', 'X'])   #PREGUNTAR A GUS
plt.xlabel("Decil")
plt.ylabel("Ingreso promedio")
plt.title("Ingreso promedio por decil") #loc="left")

#Para salvar
plt.savefig('Ingreso promedio por decil.png', dpi=96)
```

# Dispersion

```python
estados = pd.read_csv("estados.csv", encoding="latin-1")
```

```python
estados
```

```python
#Hacer el scatterplot

sns.scatterplot(x="Población_con_ingreso_inferior_a_la_línea_de_pobreza_por_ingresos", y="Rezago_educativo", hue="Año", 
                data=estados, palette="Set2", size="Año", alpha=0.8).set_title("Pobreza y Rezago Educativo")
```

```python
sns.scatterplot(x="Población_con_ingreso_inferior_a_la_línea_de_pobreza_por_ingresos", y="Carencia_por_acceso_a_los_servicios_de_salud", hue="Año", 
                data=estados, palette="Paired", size="Año", alpha=0.8).set_title("Pobreza y Sin Acceso a Salud")
```

```python
sns.scatterplot(x="Población_con_ingreso_inferior_a_la_línea_de_pobreza_por_ingresos", y="Carencia_por_acceso_a_la_seguridad_social", hue="Año", 
                data=estados, palette="Greens", size="Año", alpha=0.8).set_title("Pobreza y Sin Seguridad Social")
```

```python
sns.scatterplot(x="Población_con_ingreso_inferior_a_la_línea_de_pobreza_por_ingresos", y="Carencia_por_acceso_a_los_servicios_básicos_en_la_vivienda", hue="Año", 
                data=estados, size="Año", palette="Blues", alpha=0.8).set_title("Pobreza y Sin Servicios Básicos en Vivienda")
```

```python
#Ahora las juntamos ahorita en uno solo

f, ([ax1, ax2], [ax3, ax4]) = plt.subplots(2, 2, figsize=(10,10), sharex=True)  #sharex hace que compartan eje x

sns.scatterplot(x="Población_con_ingreso_inferior_a_la_línea_de_pobreza_por_ingresos", y="Rezago_educativo", hue="Año", 
                data=estados, palette="Set2", size="Año", alpha=0.8, ax= ax1)
ax1.set_title("Pobreza y Rezago Educativo")
ax1.set_xlabel("Pobreza")
ax1.set_ylabel("Rezago Educativo")

sns.scatterplot(x="Población_con_ingreso_inferior_a_la_línea_de_pobreza_por_ingresos", y="Carencia_por_acceso_a_los_servicios_de_salud", hue="Año", 
                data=estados, palette="Paired", size="Año", alpha=0.8, ax=ax2)
ax2.set_title("Pobreza y Sin Acceso a Salud")
ax2.set_xlabel("Pobreza")
ax2.set_ylabel("Sin servicios salud")

sns.scatterplot(x="Población_con_ingreso_inferior_a_la_línea_de_pobreza_por_ingresos", y="Carencia_por_acceso_a_la_seguridad_social", hue="Año", 
                data=estados, palette="Greens", size="Año", alpha=0.8, ax=ax3)
ax3.set_title("Pobreza y Sin Seguridad Social")
ax3.set_xlabel("Pobreza")
ax3.set_ylabel("Sin seguridad social")

sns.scatterplot(x="Población_con_ingreso_inferior_a_la_línea_de_pobreza_por_ingresos", y="Carencia_por_acceso_a_los_servicios_básicos_en_la_vivienda", hue="Año", 
                data=estados, size="Año", palette="Blues", alpha=0.8, ax= ax4)
ax4.set_title("Pobreza y Sin Servicios Básicos en Vivienda")
ax4.set_xlabel("Pobreza")
ax4.set_ylabel("Sin servicios en vivienda")

plt.show()

plt.savefig('DispersionesPobreza.png', dpi=96)
```
