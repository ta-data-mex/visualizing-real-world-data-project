```python
import pandas as pandito
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datos = pandito.read_csv('findb.csv')
```

```python
datos['Cuatrimestre'] = datos['Grupo'].apply(lambda x: x[0])
```

```python
carrera1 = datos[['Licenciatura', '1ero', 'Promedio']]
carreradf = datos.groupby(['Licenciatura'], as_index = False)['1ero'].mean()
carreradf.to_csv("primerparcialcarrera.csv", index = False)
carreradf
```

```python
sns.set()
sns.distplot(carreradf['1ero'], bins= 5)
carreradf.plot.bar()
```

```python
carrera2 = datos.groupby(['Licenciatura'], as_index = False)['2do'].mean()
carrera2.to_csv("segundoparcialcarrera.csv", index = False)
carrera2
```

```python
plt.pie(carrera2['2do'], labels=carrera2["Licenciatura"])
```

```python
carrera3 = datos.groupby(['Licenciatura'], as_index = False)['3ero'].mean()
carrera3.to_csv("Tercerparcialcarrera.csv", index = False)
carrera3
```

```python
general = pandito.merge(carrera1, pandito.merge(carrera2, carrera3))
```

```python
general['Promedio'] = (datos['1ero'] + datos['2do'] + datos['3ero'])/3
general.to_csv("PromedioCuatrimestre.csv", index = False)
general['Licenciatura'] = datos['Licenciatura']
general
```

```python
sns.scatterplot(general['1ero'] ,general['Promedio'])
```
