En la siguiente gráfica hago una categorización de los motivos de la interrupción legal del embarazo. Sin embargo, el conteo de la mayoría de las categorías es muy baja. Cabe aclarar que la base de datos contiene 58373 datos. 

```python
interrupcion.motiles.value_counts()
```

```python
sns.set(context = 'talk', style = 'darkgrid', rc={'figure.figsize':(12, 6)})
ax = sns.countplot(x = 'motiles', data = interrupcion, alpha = 0.8)
ax.set_title("Conteo motivos de interrupcion legal del embarazo")
ax.set(xlabel='motivos interrupción', ylabel='conteo')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90);
```

Interrupción voluntaria es la opción más colocada pero no es informativa al igual que la ausencia de información, que hacen las categorías sean difíciles de comparar. 

```python
motivos_sub = interrupcion[(interrupcion['motiles'] != 'interrupción voluntaria') &
                           (interrupcion['motiles'] != 'NA')]
```

```python
sns.set(context = 'talk', style = 'darkgrid', rc={'figure.figsize':(12, 6)})
ax = sns.countplot(x = 'motiles', data = motivos_sub, alpha = 0.8)
ax.set_title("Conteo motivos de interrupcion legal del embarazo")
ax.set(xlabel='motivos interrupción', ylabel='conteo')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90);
```

```python
interrupcion.year.value_counts()
```

```python
# interrupcion.pivot_table(index = 'mes', columns = 'year', aggfunc = 'count')

total_mes = interrupcion.groupby(['year', 'mes'], as_index = False).agg({'edad': 'count'})#.plot.bar()

ax = sns.barplot(x = 'year', y = 'edad', hue = 'mes', data = total_mes, palette = 'deep')
ax.set_title("Número total de interrupciones por mes y año")
ax.set(xlabel='año:mes', ylabel='total')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90);
```

```python
sub_edu_nile = interrupcion[(interrupcion.nivel_edu != 'NA') & (interrupcion.nile != 'NA')]
```

```python
sub_edu_nile.nile.value_counts()
```

```python
sub_edu_nile.nile = sub_edu_nile.nile.astype('int32')


sns.set(context = 'notebook', style = 'darkgrid', rc={'figure.figsize':(12, 10)})
ax = sns.scatterplot(x = 'nivel_edu', y = 'nile', hue = 'nivel_edu', data = sub_edu_nile, palette = 'Paired')
ax.set_title("Relación número de interrupciones anteriores con nivel educativo")
ax.set(xlabel='nivel de escolaridad', ylabel='número de interrupciones')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90);
```
